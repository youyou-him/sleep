```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

def plot_learning_curve(estimator, X, y, cv=5, scoring='accuracy'):
    train_sizes, train_scores, val_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=scoring, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
    )
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)
    
    plt.figure(figsize=(8,6))
    plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    plt.plot(train_sizes, val_mean, 'o-', color='green', label='Validation score')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='green')
    
    plt.title('Learning Curve')
    plt.xlabel('Training Examples')
    plt.ylabel(scoring.capitalize())
    plt.legend(loc='best')
    plt.grid()
    plt.show()

# 데이터 분리 (X, y는 미리 준비된 상태여야 함)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 모델 생성 및 학습
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 학습 곡선 출력
plot_learning_curve(model, X_train, y_train, cv=3)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy')
print(f"Cross-validation scores: {scores}")
print(f"Mean CV accuracy: {scores.mean():.4f}, Std: {scores.std():.4f}")
```

## 군집별 맞춤 수면 개선 메시지 작성 코드

```python
# 군집별 맞춤 수면 개선 추천 메시지 함수 정의
def recommend_sleep(cluster_id):
    if cluster_id == 0:
        return "스트레스가 낮고 활동량이 적당한 건강한 군집입니다. 꾸준한 운동과 규칙적 수면 유지가 좋습니다."
    elif cluster_id == 1:
        return "스트레스가 매우 높고 활동량이 높은 군집입니다. 스트레스 관리와 휴식 시간이 필요합니다."
    elif cluster_id == 2:
        return "중간 정도 활동과 스트레스 수준의 군집입니다. 수면 환경 개선과 적절한 신체 활동 권장합니다."
    else:
        return "알 수 없는 군집입니다."

# 군집별 추천 메시지 컬럼 추가
df_cluster['Recommendation'] = df_cluster['Cluster'].apply(recommend_sleep)

# 군집별 추천 메시지 확인
for cluster_id in sorted(df_cluster['Cluster'].unique()):
    print(f"군집 {cluster_id} 추천 메시지:")
    print(recommend_sleep(cluster_id))
    print("-" * 40)
```

## 집별 수면 질(Quality of Sleep) 관계 분석 및 시각화

```python
# 군집화에 쓸 컬럼만 명확히 지정!
cluster_cols = ['Gender', 'Occupation', 'BMI Category', 'Age', 'Heart Rate', 'Stress Level']  # 반드시 숫자형만

X_cluster = df_cluster[cluster_cols].copy()  # 이때, cluster_cols 안에 문자열 설명 컬럼이 있으면 안됨

# (혹시라도 위 컬럼들에 object 타입이 있으면 LabelEncoder로 인코딩)
from sklearn.preprocessing import LabelEncoder
for col in ['Gender', 'Occupation', 'BMI Category']:
    if X_cluster[col].dtype == 'object':
        le = LabelEncoder()
        X_cluster[col] = le.fit_transform(X_cluster[col])

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster['Cluster'] = kmeans.fit_predict(X_cluster)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# --- 1. 데이터 준비 ---
# 예시: 원본 데이터프레임 df에서 군집에 사용할 컬럼만 추출해 df_cluster 생성
# 필요한 컬럼명 리스트 작성 (예: Person ID, Quality of Sleep 등 제외)
use_cols = [
    'Gender', 'Age', 'Occupation', 'Physical Activity Level', 'Stress Level',
    'BMI Category', 'Heart Rate', 'Systolic', 'Diastolic', 'risk_score', 'gubun'
]

# df가 이미 로드되어 있다고 가정
df_cluster = df[use_cols].copy()

# --- 2. 범주형 변수 인코딩 ---
cat_cols = df_cluster.select_dtypes(include=['object', 'category']).columns

le_dict = {}
for col in cat_cols:
    le = LabelEncoder()
    df_cluster[col] = le.fit_transform(df_cluster[col].astype(str))
    le_dict[col] = le  # 필요 시 디코딩용 저장

# --- 3. 군집용 피처 선택 (정답, ID 제외는 이미 제외했으므로 그대로 사용) ---
X_cluster = df_cluster.copy()

# --- 4. KMeans 군집화 ---
k = 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df_cluster['Cluster'] = kmeans.fit_predict(X_cluster)

# --- 5. 군집별 수치형 변수 평균값 계산 ---
numeric_cols = df_cluster.select_dtypes(include=[np.number]).columns.drop('Cluster')
cluster_summary_numeric = df_cluster.groupby('Cluster')[numeric_cols].mean()
print("\n[군집별 수치형 변수 평균]")
display(cluster_summary_numeric)

# --- 6. 군집별 범주형 변수 최빈값(mode) 계산 ---
cat_cols = cat_cols.intersection(df_cluster.columns)  # 군집 포함된 범주형만

cluster_summary_categorical = pd.DataFrame()
for col in cat_cols:
    mode_indices = df_cluster.groupby('Cluster')[col].agg(lambda x: x.mode().iloc[0])
    mode_values = mode_indices.map(lambda x: le_dict[col].inverse_transform([x])[0])
    cluster_summary_categorical[col] = mode_values

print("\n[군집별 범주형 변수 최빈값]")
display(cluster_summary_categorical)

# --- 7. 군집별 주요 변수 시각화 (한 줄에 2개씩) ---
plot_pairs = [
    ('Gender', 'Age'),
    ('Physical Activity Level', 'Stress Level'),
    ('BMI Category', 'Heart Rate'),
    ('Systolic', 'Diastolic'),
    ('Physical Activity Level', 'Heart Rate'),
]

cols = 2
rows = (len(plot_pairs) + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(cols * 7, rows * 5))
axes = axes.flatten()

for i, (x_feat, y_feat) in enumerate(plot_pairs):
    sns.scatterplot(x=x_feat, y=y_feat, data=df_cluster, hue='Cluster', palette='Set1', ax=axes[i])
    axes[i].set_title(f'{x_feat} vs {y_feat} by Cluster')
    axes[i].legend(loc='best')

# 남는 subplot 삭제
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 예시: 군집별 주요 변수 평균 (이미 cluster 컬럼이 포함된 df_cluster 사용)
cluster_summary = df_cluster.groupby('Cluster').mean()

# 군집별 특성 해석 및 맞춤형 수면 개선 추천 문구 작성 함수
def recommend_sleep(cluster_id):
    if cluster_id == 0:
        return "스트레스가 낮고 활동량이 적당한 건강한 군집입니다. 꾸준한 운동과 규칙적 수면 유지가 좋습니다."
    elif cluster_id == 1:
        return "스트레스가 매우 높고 활동량이 높은 군집입니다. 스트레스 관리와 휴식 시간이 필요합니다."
    elif cluster_id == 2:
        return "중간 정도 활동과 스트레스 수준의 군집입니다. 수면 환경 개선과 적절한 신체 활동 권장합니다."
    else:
        return "알 수 없는 군집입니다."

# 각 군집별 요약 + 추천 문구 출력
for cluster_id, row in cluster_summary.iterrows():
    print(f"군집 {cluster_id} 특성:")
    print(row.round(2))
    print("추천:", recommend_sleep(cluster_id))
    print("-" * 50)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

range_n_clusters = list(range(2, 10))
inertia = []
silhouette = []

for k in range_n_clusters:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = km.fit_predict(X_cluster)
    inertia.append(km.inertia_)
    silhouette.append(silhouette_score(X_cluster, cluster_labels))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(range_n_clusters, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')

plt.subplot(1,2,2)
plt.plot(range_n_clusters, silhouette, 'ro-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis For Optimal k')

plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def plot_silhouette_scores(X, cluster_range):
    scores = []
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)
        score = silhouette_score(X, labels)
        scores.append(score)
    
    plt.figure(figsize=(8,5))
    plt.plot(cluster_range, scores, marker='o')
    plt.title('Silhouette Scores for Different k')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.grid(True)
    plt.show()
    
    return scores

# 예시 실행
k_values = range(2, 11)
sil_scores = plot_silhouette_scores(X_cluster, k_values)
```
