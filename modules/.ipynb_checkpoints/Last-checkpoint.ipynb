### 'Occupation' 컬럼의 ANOVA 분석

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA를 위한 데이터 준비
# 'Quality_of_Sleep'을 종속 변수로, 'Occupation'을 독립 변수로 하는 모델 정의
# statsmodels의 ols 함수는 formula API를 사용하여 모델을 쉽게 정의할 수 있습니다.
# 'Occupation' 컬럼의 데이터 타입이 object이므로, 문자열로 처리되도록 ensure_ascii=False를 사용합니다.
# 또는 'Occupation' 컬럼을 category 타입으로 변환하여 사용할 수도 있습니다. 여기서는 간단히 문자열로 사용합니다.

# 모델 정의 (Quality_of_Sleep ~ C(Occupation) 형태로 지정하여 Occupation을 범주형으로 처리)
model = ols('Quality_of_Sleep ~ C(Occupation)', data=df).fit()

# ANOVA 테이블 출력
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

p-value가 0에 매우 가까움. 따라서 'Occupation'(직업) 범주별로 'Quality_of_Sleep'(수면의 질)의 평균에 통계적으로 유의미한 차이가 있다고 결론

### 'BMI_Category' 컬럼의 ANOVA 분석

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA를 위한 데이터 준비
# 'Quality_of_Sleep'을 종속 변수로, 'BMI_Category'를 독립 변수로 하는 모델 정의
# BMI_Category 컬럼은 이미 숫자형이므로 C()로 감싸 범주형으로 처리합니다.

# 모델 정의 (Quality_of_Sleep ~ C(BMI_Category) 형태로 지정하여 BMI_Category를 범주형으로 처리)
model = ols('Quality_of_Sleep ~ C(BMI_Category)', data=df).fit()

# ANOVA 테이블 출력
anova_table_bmi = sm.stats.anova_lm(model, typ=2)

print(anova_table_bmi)
```

p-value가 0에 매우 가까움. 따라서 'Occupation'(직업) 범주별로 'Quality_of_Sleep'(수면의 질)의 평균에 통계적으로 유의미한 차이가 있다고 결론

## 다시 데이터 분할

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

# 인코딩을 시작할 데이터 프레임을 복사하여 준비합니다.
# (원본 df를 사용하거나, 다른 전처리가 끝난 df_copy를 사용할 수 있습니다)
df_processed_encoded = df.copy() 

# 인코딩할 범주형 컬럼들의 리스트
# 실제 데이터에 맞게 컬럼 이름을 확인하고 수정하세요.
categorical_cols = ['Gender', 'Occupation', 'BMI_Category'] 

# 각 범주형 컬럼에 대해 LabelEncoder를 적용하는 반복문
for col in categorical_cols:
    le = LabelEncoder()
    df_processed_encoded[col] = le.fit_transform(df_processed_encoded[col])

# 인코딩이 잘 되었는지 상위 5개 데이터와 정보 확인
print("--- 인코딩 후 데이터 샘플 ---")
print(df_processed_encoded.head())
print("\n--- 인코딩 후 데이터 정보 ---")
df_processed_encoded.info()

# 특성 변수 (X_final)와 목표 변수 (y_final) 정의 (df_processed_encoded 사용)
X_final = df_processed_encoded.drop('Quality_of_Sleep', axis=1)
y_final = df_processed_encoded['Quality_of_Sleep']

# 전처리 완료된 데이터를 훈련 세트와 테스트 세트로 분할
# 이전과 동일한 test_size와 random_state 사용
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X_final, y_final, test_size=0.2, random_state=42)

print("최종 훈련 세트 특성 (X_train_final) 형태:", X_train_final.shape)
print("최종 테스트 세트 특성 (X_test_final) 형태:", X_test_final.shape)
print("최종 훈련 세트 목표 변수 (y_train_final) 형태:", y_train_final.shape)
print("최종 테스트 세트 목표 변수 (y_test_final) 형태:", y_test_final.shape)
```

## 스케일링
* 분류모델에서는 스케일링이 필수X
* 군집분석에서는 거리 기반이라서 무조건 스케일링 필요.(안 하면 왜곡이 심함)

```python
from sklearn.preprocessing import StandardScaler

# 스케일링할 수치형 컬럼 선택 (원-핫 인코딩된 컬럼 제외)
numerical_cols_final = ['Age', 'Sleep_Duration', 'Stress_Level', 'BMI_Category', 'Heart_Rate', 'MAP']

# StandardScaler 초기화
scaler = StandardScaler()

# 훈련 세트의 수치형 컬럼에 대해 스케일러 학습 및 변환
X_train_final[numerical_cols_final] = scaler.fit_transform(X_train_final[numerical_cols_final])

# 테스트 세트의 수치형 컬럼 변환 (훈련 세트에서 학습된 스케일러 사용)
X_test_final[numerical_cols_final] = scaler.transform(X_test_final[numerical_cols_final])

print("스케일링 후 훈련 세트 특성 (X_train_final) head:")
display(X_train_final.head())

print("\n스케일링 후 테스트 세트 특성 (X_test_final) head:")
display(X_test_final.head())
```

```python
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier # Import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold

# 사용할 모델 정의 (랜덤 포레스트 - 이전 파이프라인에서 사용된 설정 활용)
model = RandomForestClassifier(random_state=42)

# SMOTE와 모델을 포함하는 파이프라인 생성
# 학습 곡선 계산 시 작은 훈련 세트 크기에서도 SMOTE가 작동하도록 k_neighbors를 조정합니다.
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42, k_neighbors=1)), # Set k_neighbors to a small value like 1
    ('classifier', model)
])

# Stratified K-Fold 교차 검증 설정 (학습 곡선 계산 내부에서 사용)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 학습 곡선 계산
train_sizes, train_scores, validation_scores = learning_curve(
    estimator=pipeline,
    X=X_train_final,
    y=y_train_final,
    cv=cv,
    train_sizes=np.linspace(0.1, 1.0, 10), # 훈련 세트 크기를 10단계로 나눔
    scoring='accuracy',
    n_jobs=-1 # 사용할 CPU 코어 수
)

# 훈련 점수와 검증 점수의 평균 및 표준편차 계산
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
validation_scores_mean = np.mean(validation_scores, axis=1)
validation_scores_std = np.std(validation_scores, axis=1)

# 학습 곡선 시각화
plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, validation_scores_mean - validation_scores_std,
                 validation_scores_mean + validation_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, validation_scores_mean, 'o-', color="g", label="Cross-validation score")

plt.title("Learning Curves (Random Forest with SMOTE)")
plt.xlabel("Training Examples")
plt.ylabel("Accuracy Score")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

## 엘보차트 그리기

```python
# [설명]: K-Means에서 최적의 군집 개수(k)를 찾기 위한 엘보 차트를 그립니다.

# 군집 개수(k) 범위 설정
max_k = 10 # 테스트할 최대 군집 개수
sse = [] # 각 k에 대한 SSE(Sum of Squared Errors) 값을 저장할 리스트

# 각 k 값에 대해 K-Means 수행 및 SSE 계산
for k in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X_cluster_scaled) # 스케일링된 데이터 사용
    sse.append(kmeans.inertia_) # inertia_ 속성이 SSE 값입니다.

# 엘보 차트 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_k + 1), sse, marker='o', linestyle='--')
plt.title('Elbow Method', fontsize=15)
plt.xlabel('k', fontsize=12)
plt.ylabel('SSE (Sum of Squared Errors / Inertia)', fontsize=12)
plt.xticks(range(1, max_k + 1)) # x축 눈금 설정
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n엘보 차트가 생성되었습니다. SSE 값이 급격히 줄어들다가 완만해지는 '팔꿈치' 지점을 찾아보세요.")
```

## 다양한 K 값에 따른 2D 군집 산점도 시각화

```python
# [설명]: k=3, 4, 5일 때 각각 PCA를 이용한 2차원 산점도를 그려 군집 응집도를 시각적으로 비교합니다.
from sklearn.decomposition import PCA # PCA 임포트 (상단에 이미 임포트되어 있어야 함)

# 테스트할 k 값 리스트
k_values_to_plot = [3, 4, 5]

# 각 k 값에 대해 산점도 그리기
for k in k_values_to_plot:
    # 1. K-Means 군집화 수행 (스케일링된 데이터 X_cluster_scaled 사용)
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    cluster_labels = kmeans.fit_predict(X_cluster_scaled)

    # 2. 실루엣 점수 계산 (응집도 지표)
    silhouette_avg = silhouette_score(X_cluster_scaled, cluster_labels)

    # 3. PCA를 이용해 2차원으로 차원 축소 (군집화에 사용했던 스케일링된 데이터 사용)
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_cluster_scaled)

    # 4. PCA 결과와 군집 라벨을 포함하는 데이터프레임 생성
    # X_pca의 첫 번째 주성분을 'PC1', 두 번째 주성분을 'PC2'로 이름 붙입니다.
    # 'Cluster' 정보는 해당 k값의 cluster_labels를 사용합니다.
    pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
    pca_df['Cluster'] = cluster_labels # 해당 k값의 클러스터 라벨을 pca_df에 추가

    # 5. 군집별 산점도 시각화
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df,
                    palette='viridis', s=100, alpha=0.7,
                    legend='full') # 'legend=full'로 범례를 명확하게 표시

    plt.title(f'PCA 2D Scatter: K={k} 군집별 분포 (Silhouette Score: {silhouette_avg:.3f})', fontsize=12)
    plt.xlabel('주성분 1 (PC1)', fontsize=9)
    plt.ylabel('주성분 2 (PC2)', fontsize=9)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show() # 차트를 주피터에 표시

print("\nK=3, 4, 5일 때의 PCA 2D 산점도가 각각 생성되었습니다. 각 군집의 응집도와 분리도를 시각적으로 비교해보세요.")
```

## 군집 분석 스케일링에 사용된 StandardScaler 객체 저장

군집 분석 전 데이터 스케일링에 사용되었던 `StandardScaler` 객체를 `scaler.pkl` 파일로 저장합니다. 이 스케일러는 군집 분석 모델과 별개로 저장되어, 필요에 따라 동일한 스케일링 변환을 적용할 수 있게 합니다.

## 최종 범주형 피처 매핑 딕셔너리 생성 및 저장 (kaggle_sleep.csv 기반)

원본 데이터셋 (`kaggle_sleep.csv`)에서 확인된 범주형 피처들의 문자열 레이블과 전처리 과정에서 사용된 숫자 값 사이의 매핑 정보를 담은 딕셔너리를 생성하여 `le_dict.pkl` 파일로 저장합니다. 이 딕셔너리는 웹 서비스에서 사용자 입력 변환 및 결과 역변환에 사용됩니다.

```python
import joblib
import pandas as pd # 혹시 필요할 경우를 대비하여 임포트

# 매핑 정보 정의 (원본 데이터 및 전처리 과정 기반)

# 'Gender' 매핑 (0=Female, 1=Male로 추정)
gender_mapping = {0: 'Female', 1: 'Male'}
gender_inverse_mapping = {'Female': 0, 'Male': 1}

# 'Occupation' 매핑 (원본 문자열 -> 변환된 숫자)
# 원본: '1', '2', '3', '4', 'Other' -> 변환: 1, 2, 3, 4, 5
occupation_mapping = {
    1: '1', 2: '2', 3: '3', 4: '4', 5: 'Other'
}
occupation_inverse_mapping = {
    '1': 1, '2': 2, '3': 3, '4': 4, 'Other': 5
}

# 'BMI_Category' 매핑 (원본 문자열 -> 전처리 숫자 1, 2, 3)
# 원본: 'Normal', 'Normal Weight', 'Overweight', 'Obese'
# 전처리: 1, 2, 3 (Normal/Normal Weight -> 1, Overweight -> 2, Obese -> 3 으로 추정)
bmi_category_mapping = {
    1: 'Normal/Normal Weight', # 숫자 1에 해당하는 원본 레이블
    2: 'Overweight',           # 숫자 2에 해당하는 원본 레이블
    3: 'Obese'               # 숫자 3에 해당하는 원본 레이블
}
# 역 매핑 (웹 서비스 입력용: 원본 문자열 -> 숫자)
bmi_category_inverse_mapping = {
    'Normal': 1,
    'Normal Weight': 1,
    'Overweight': 2,
    'Obese': 3
}


# 'Sleep Disorder' 매핑 (원본 문자열 -> 숫자 0, 1, 2)
# 원본: 'None', 'Insomnia', 'Sleep Apnea', NaN
# 매핑: 'None' -> 0, 'Insomnia' -> 1, 'Sleep Apnea' -> 2 (NaN 제외)
sleep_disorder_mapping = {
    0: 'None',
    1: 'Insomnia',
    2: 'Sleep Apnea'
}
sleep_disorder_inverse_mapping = {
    'None': 0,
    'Insomnia': 1,
    'Sleep Apnea': 2
}


# 전체 매핑 딕셔너리 통합

# 숫자 -> 문자열 매핑 (결과 표시용)
label_to_string_mapping = {
    'Gender': gender_mapping,
    'Occupation': occupation_mapping,
    'BMI_Category': bmi_category_mapping,
    'Sleep Disorder': sleep_disorder_mapping # 컬럼 이름 일치 ('Sleep Disorder'로 저장)
}

# 문자열 -> 숫자 매핑 (웹 서비스 입력 변환용)
string_to_label_mapping = {
    'Gender': gender_inverse_mapping,
    'Occupation': occupation_inverse_mapping,
    'BMI_Category': bmi_category_inverse_mapping,
    'Sleep Disorder': sleep_disorder_inverse_mapping # 컬럼 이름 일치
}

# 두 딕셔너리를 포함하는 최종 딕셔너리
mapping_dict = {
    'label_to_string': label_to_string_mapping,
    'string_to_label': string_to_label_mapping
}


# 딕셔너리 저장할 파일 경로 설정
output_dict_path = '/content/le_dict.pkl'

# 딕셔너리 저장
try:
    joblib.dump(mapping_dict, output_dict_path)
    print(f"최종 범주형 피처 매핑 딕셔너리를 '{output_dict_path}' 파일로 성공적으로 저장했습니다.")
    print("이 파일은 웹 서비스에서 사용자 입력 변환 및 결과 역변환에 사용됩니다.")
except Exception as e:
    print(f"매핑 딕셔너리 저장 중 오류 발생: {e}")

# 저장된 딕셔너리 내용 확인 (선택 사항)
# print("\n저장된 딕셔너리 내용:")
# display(mapping_dict)
```

## 군집별 주요 피처 평균값 딕셔너리 저장

각 군집(`Cluster`)별로 주요 수치형 피처들의 평균 값을 계산하여 딕셔너리에 저장하고 `.pkl` 파일로 저장합니다. 이 딕셔너리는 웹 서비스에서 사용자 지표와 군집 평균을 비교하는 시각화에 사용됩니다.

```python
import joblib
import pandas as pd

# df 데이터프레임에 'Cluster' 컬럼이 추가되어 있다고 가정합니다.
# 만약 df가 정의되지 않았거나 'Cluster' 컬럼이 없다면, 데이터를 다시 로드하고 군집 분석을 수행해야 합니다.
# 현재 런타임에는 df가 정의되어 있고 Cluster 컬럼이 있습니다.

# 군집별 평균을 계산할 주요 수치형 피처 목록 정의
# 모든 수치형 컬럼을 포함하거나, 웹 서비스에서 비교할 특정 컬럼만 선택할 수 있습니다.
# 여기서는 군집 분석 시 사용되었던 수치형 특성들과 관련 있다고 판단되는 컬럼들을 포함합니다.
# 'MAP' 컬럼이 df에 있는지 확인하고, 없다면 계산합니다.
if 'MAP' not in df.columns:
    if 'Systolic' in df.columns and 'Diastolic' in df.columns:
        df['MAP'] = df['Diastolic'] + (1/3) * (df['Systolic'] - df['Diastolic'])
        print("MAP 컬럼을 df에 추가했습니다.")
    else:
        print("경고: MAP 계산에 필요한 'Systolic' 또는 'Diastolic' 컬럼이 df에 없습니다.")

# 군집별 평균 계산 대상 컬럼 (df에 존재하는 수치형 컬럼 중 'Cluster' 제외)
# 'Quality_of_Sleep'도 평균을 볼 수 있지만, 비교 대상 지표로 사용할지 여부에 따라 포함 여부 결정
features_for_average = [col for col in df.select_dtypes(include=np.number).columns if col != 'Cluster']
# 'Quality_of_Sleep'을 포함할지 여부는 결정에 따라 조정 가능

print(f"군집별 평균 계산 대상 컬럼: {features_for_average}")


# 군집별 평균 계산
# numeric_only=True를 사용하여 숫자형 컬럼만 선택하여 평균 계산
cluster_averages_df = df.groupby('Cluster')[features_for_average].mean()

# 결과를 파이썬 딕셔너리 형태로 변환
# { cluster_id: { feature_name: average_value, ... }, ... } 형태
cluster_averages_dict = cluster_averages_df.to_dict(orient='index')

print("\n계산된 군집별 평균 딕셔너리 (일부):")
# 딕셔너리의 첫 번째 항목만 출력하여 확인
for cluster_id, averages in list(cluster_averages_dict.items())[:1]:
    print(f"Cluster {cluster_id}: {list(averages.items())[:5]}...") # 처음 5개 항목만 표시
# print(cluster_averages_dict) # 전체 딕셔너리 출력 (선택 사항)


# 딕셔너리 저장할 파일 경로 설정
output_dict_path = '/content/cluster_averages.pkl'

# 딕셔너리 저장
try:
    joblib.dump(cluster_averages_dict, output_dict_path)
    print(f"\n군집별 평균값 딕셔너리를 '{output_dict_path}' 파일로 성공적으로 저장했습니다.")
    print("이 파일은 웹 서비스에서 사용자 지표와 군집 평균 비교 차트를 그릴 때 사용될 수 있습니다.")
except Exception as e:
    print(f"군집별 평균값 딕셔너리 저장 중 오류 발생: {e}")
```

```python
import pandas as pd
import joblib
import numpy as np # numpy 임포트 추가
from sklearn.preprocessing import StandardScaler # StandardScaler 클래스 임포트
from sklearn.cluster import KMeans # KMeans 클래스 임포트

# df 데이터프레임 다시 로드 (Cluster 컬럼이 포함된 상태로 추정)
# 군집 분석 후 df에 Cluster 컬럼이 추가되었으므로, 해당 상태를 로드하거나 재생성해야 합니다.
# 여기서는 Cluster 컬럼이 추가된 상태를 가정하고, MAP 컬럼이 없을 경우 다시 계산합니다.
# 만약 Cluster 컬럼이 로드되지 않는다면, 군집 분석 부분을 다시 실행해야 합니다.

# 이전에 사용했던 파일 경로 중 Cluster 컬럼이 추가되었을 가능성이 있는 파일을 로드합니다.
# 만약 final_ori_df.csv에 Cluster 컬럼이 없다면, 군집 분석 코드를 다시 실행해야 합니다.
file_path_with_cluster = '/content/final_ori_df.csv' # Cluster 컬럼이 저장되었을 수도 있는 파일

try:
    df = pd.read_csv(file_path_with_cluster)
    print(f"데이터를 성공적으로 로드했습니다: {file_path_with_cluster}")
    display(df.head())

    # Cluster 컬럼이 로드되었는지 확인
    if 'Cluster' not in df.columns:
        print("경고: 로드된 데이터에 'Cluster' 컬럼이 없습니다. 군집 분석을 다시 수행하여 'Cluster' 컬럼을 추가합니다.")
        # 군집 분석 관련 변수들이 정의되어 있어야 합니다.
        # X_cluster, label_encoded_categorical_cols, scaler, kmeans 객체 등이 필요합니다.
        # 이 부분은 이전 셀들을 다시 실행해야 할 수 있습니다.
        # 여기서는 Cluster 컬럼이 이미 있다고 가정하고 진행합니다.
        # 실제 웹 서비스에서는 Cluster 컬umn이 미리 추가된 데이터를 사용하거나,
        # 웹 서비스 내부에서 군집 분석을 수행해야 합니다.
        # 노트북 실행 순서 상 Cluster 컬럼은 메모리 상의 df에 추가되었으나 파일로 저장되지는 않았을 수 있습니다.
        # 따라서 여기서는 MAP 계산 후 군집 분석을 다시 수행하여 Cluster 컬럼을 추가하는 코드를 포함합니다.

        # MAP 컬럼이 없다면 추가 (군집 분석에 필요)
        if 'MAP' not in df.columns:
             if 'Systolic' in df.columns and 'Diastolic' in df.columns:
                df['MAP'] = df['Diastolic'] + (1/3) * (df['Systolic'] - df['Diastolic'])
                print("MAP 컬럼을 df에 추가했습니다.")
             else:
                print("경고: MAP 계산에 필요한 'Systolic' 또는 'Diastolic' 컬럼이 없습니다.")

        # 군집 분석 다시 수행하여 Cluster 컬럼 추가
        # 주의: 이 부분은 이전 군집 분석 코드를 복사한 것입니다.
        # scaler와 kmeans 객체는 이 코드를 통해 다시 생성됩니다.
        print("Cluster 컬럼 추가를 위해 군집 분석 다시 수행...")
        # 군집 분석 특성 데이터 준비 (df 사용)
        # MAP 계산 전에 Systolic, Diastolic이 필요하므로 drop 목록에서 제외
        # Occupation은 여기서 object일 수 있으므로 변환 로직 필요
        X_cluster_temp = df.drop(columns=['Sleep_Duration', 'Quality_of_Sleep', 'gubun', 'Sleep Disorder'], errors='ignore')

        # MAP 계산 (X_cluster_temp에서 수행)
        if 'MAP' not in X_cluster_temp.columns and 'Systolic' in X_cluster_temp.columns and 'Diastolic' in X_cluster_temp.columns:
             X_cluster_temp['MAP'] = X_cluster_temp['Diastolic'] + (1/3) * (X_cluster_temp['Systolic'] - X_cluster_temp['Diastolic'])
             X_cluster_temp = X_cluster_temp.drop(['Systolic', 'Diastolic'], axis=1) # MAP 계산 후 원본 삭제
             print("X_cluster_temp에 MAP 컬럼 추가 및 Systolic, Diastolic 삭제.")
        elif 'MAP' not in X_cluster_temp.columns:
             print("경고: X_cluster_temp에 MAP 계산에 필요한 Systolic/Diastolic 컬럼이 없습니다. MAP 계산을 건너뜀.")


        # Occupation 변환 (필요하다면) - X_cluster_temp에 적용
        if 'Occupation' in X_cluster_temp.columns and X_cluster_temp['Occupation'].dtype == 'object':
            # 이전에 Other를 5로 바꿨던 수동 변환을 여기서 다시 적용
            X_cluster_temp['Occupation'] = X_cluster_temp['Occupation'].replace('Other', 5).astype(int)
            print("X_cluster_temp Occupation 컬럼 숫자형으로 변환.")


        # 스케일링 (새로운 scaler 객체 사용) - X_cluster_temp에 적용
        # label_encoded_categorical_cols 변수가 정의되어 있어야 합니다.
        if 'label_encoded_categorical_cols' not in locals():
             label_encoded_categorical_cols = ['Gender', 'Occupation', 'BMI_Category'] # 다시 정의 (이전 수정 반영)
             print("label_encoded_categorical_cols 다시 정의됨.")


        # 스케일링 대상 컬럼 재정의 (X_cluster_temp 기반)
        numerical_cols_for_clustering_temp = [
            col for col in X_cluster_temp.columns
            if col not in label_encoded_categorical_cols and X_cluster_temp[col].dtype in [np.int64, np.float64]
        ]
        print(f"스케일링 대상 컬럼 (X_cluster_temp): {numerical_cols_for_clustering_temp}")

        scaler_temp = StandardScaler()
        # 스케일링 대상 컬럼이 비어있지 않은 경우에만 스케일링 시도
        if numerical_cols_for_clustering_temp:
             X_cluster_scaled_temp = X_cluster_temp.copy()
             X_cluster_scaled_temp[numerical_cols_for_clustering_temp] = scaler_temp.fit_transform(X_cluster_scaled_temp[numerical_cols_for_clustering_temp])
             print("군집 분석 데이터 스케일링 다시 완료.")
        else:
             X_cluster_scaled_temp = X_cluster_temp.copy() # 스케일링 대상 없으면 복사만
             print("스케일링 대상 컬럼이 없어 스케일링을 건너뜁니다.")


        # KMeans 군집 분석 (새로운 kmeans 객체 사용)
        # 스케일링된 데이터 (X_cluster_scaled_temp)를 사용합니다.
        # 이때 사용되는 컬럼은 X_cluster_scaled_temp의 모든 컬럼입니다.
        print("KMeans 군집 분석 수행 (X_cluster_scaled_temp 사용)...")
        try:
            kmeans_temp = KMeans(n_clusters=4, random_state=42, n_init='auto')
            cluster_labels = kmeans_temp.fit_predict(X_cluster_scaled_temp)
            df['Cluster'] = cluster_labels # df에 Cluster 컬럼 추가
            print("군집 분석 다시 완료 및 Cluster 컬럼 추가.")
        except Exception as kmeans_e:
             print(f"KMeans 군집 분석 중 오류 발생: {kmeans_e}")
             print("군집 분석을 건너뛰고 Cluster 컬럼을 추가하지 않습니다.")


    # Cluster 컬럼이 이제 df에 있다고 가정하고 진행합니다.

    # Cluster 컬럼이 실제로 df에 있는지 최종 확인
    if 'Cluster' not in df.columns:
        print("오류: 군집 분석 후에도 'Cluster' 컬럼이 df에 추가되지 않았습니다. 군집별 평균 계산 불가.")
    else:
        # 군집별 평균 계산 대상 컬럼 (df에 존재하는 수치형 컬럼 중 'Cluster' 제외)
        # 'Quality_of_Sleep'을 포함할지 여부는 결정에 따라 조정 가능
        features_for_average = [col for col in df.select_dtypes(include=np.number).columns if col != 'Cluster']

        print(f"\n군집별 평균 계산 대상 컬럼: {features_for_average}")

        # 군집별 평균 계산
        # numeric_only=True를 사용하여 숫자형 컬럼만 선택하여 평균 계산
        cluster_averages_df = df.groupby('Cluster')[features_for_average].mean()

        # 결과를 파이썬 딕셔너리 형태로 변환
        # { cluster_id: { feature_name: average_value, ... }, ... } 형태
        cluster_averages_dict = cluster_averages_df.to_dict(orient='index')

        print("\n계산된 군집별 평균 딕셔너리 (일부):")
        # 딕셔너리의 첫 번째 항목만 출력하여 확인
        for cluster_id, averages in list(cluster_averages_dict.items())[:1]:
            print(f"Cluster {cluster_id}: {list(averages.items())[:5]}...") # 처음 5개 항목만 표시


        # 딕셔너리 저장할 파일 경로 설정
        output_dict_path = '/content/cluster_averages.pkl'

        # 딕셔너리 저장
        try:
            joblib.dump(cluster_averages_dict, output_dict_path)
            print(f"\n군집별 평균값 딕셔너리를 '{output_dict_path}' 파일로 성공적으로 저장했습니다.")
            print("이 파일은 웹 서비스에서 사용자 지표와 군집 평균 비교 차트를 그릴 때 사용될 수 있습니다.")
        except Exception as e:
            print(f"군집별 평균값 딕셔너리 저장 중 오류 발생: {e}")

except FileNotFoundError:
    print(f"오류: 지정된 경로에 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요: {file_path_with_cluster}")
except Exception as e:
    print(f"데이터 로딩 또는 처리 중 오류 발생: {e}")
```

## MAP 피처 포함 예측 파이프라인 정의 및 저장

수면 건강 점수 예측을 위해 'Systolic'과 'Diastolic' 대신 'MAP' 피처를 포함하는 데이터로 학습된 파이프라인을 `model_pipeline_no_bp.pkl` 파일로 저장합니다.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import lightgbm as lgb
import joblib
import numpy as np # numpy 임포트 추가
import pandas as pd # pandas 임포트 추가

# 예측에 사용될 특성 컬럼 목록 정의 (MAP 포함, Systolic, Diastolic 제외)
# 'Quality_of_Sleep'은 예측할 목표 변수이므로 제외합니다.
# 'Gender', 'Age', 'Occupation', 'Sleep_Duration', 'Stress_Level', 'BMI_Category', 'Heart_Rate', 'MAP'을 포함합니다.
feature_cols_no_bp = ['Gender', 'Age', 'Occupation', 'Sleep_Duration', 'Stress_Level',
                      'BMI_Category', 'Heart_Rate', 'MAP']

# 파이프라인 학습을 위해 사용할 데이터 준비
# 현재 df 데이터프레임을 사용하되, 필요한 컬럼만 선택합니다.
# df에는 MAP 컬럼이 이미 추가되어 있습니다 (셀 3fee40d7 실행 결과).
# Systolic, Diastolic은 이 파이프라인에서 제외합니다.
X_pipeline_train_no_bp = df[feature_cols_no_bp].copy()
y_pipeline_train_no_bp = df['Quality_of_Sleep'].copy() # 예측 대상 컬럼

# Occupation 컬럼이 object 타입일 경우 int로 변환 (필요하다면)
# 현재 df에서는 Occupation이 int64 타입이므로 이 변환은 건너뛰게 됩니다.
if X_pipeline_train_no_bp['Occupation'].dtype == 'object':
    X_pipeline_train_no_bp['Occupation'] = X_pipeline_train_no_bp['Occupation'].replace('Other', 5).astype(int)
    print("X_pipeline_train_no_bp Occupation 컬럼 숫자형으로 변환.")


# 스케일링할 수치형 컬럼 목록 (파이프라인 내부에서 적용)
# 모든 특성 컬럼을 스케일링 대상으로 포함합니다.
# 'Gender', 'Occupation', 'BMI_Category'는 범주형이지만, 현재 숫자 형태이므로 스케일링 대상에 포함합니다.
numerical_features_no_bp = feature_cols_no_bp

# LightGBM 모델 정의 (이전에 학습된 모델의 설정 활용)
# num_class는 예측하려는 수면의 질 레벨의 총 개수 (1부터 9까지 총 9개 클래스)
num_classes = len(y_pipeline_train_no_bp.unique()) # y_pipeline_train_no_bp에서 고유 클래스 개수 확인
final_lgbm_model_no_bp = lgb.LGBMClassifier(objective='multiclass', num_class=num_classes, random_state=42)

# 최종 파이프라인 정의
# 1. StandardScaler: 수치형 특성 스케일링
# 2. classifier: LightGBM 모델
# Note: SMOTE는 학습 시에만 사용되므로 예측 파이프라인에는 포함하지 않습니다.
prediction_pipeline_no_bp = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', final_lgbm_model_no_bp)
])

# 파이프라인 학습
# 스케일링은 파이pe라인 내부에서 수치형 컬럼에만 자동으로 적용됩니다.
# LightGBM 모델도 파이프라인의 일부로 학습됩니다.
print("MAP 포함 예측 파이프라인 학습 시작...")
prediction_pipeline_no_bp.fit(X_pipeline_train_no_bp, y_pipeline_train_no_bp)
print("MAP 포함 예측 파이프라인 학습 완료.")


# 파이프라인 객체를 .pkl 파일로 저장
output_pipeline_path_no_bp = '/content/model_pipeline_no_bp.pkl'
try:
    joblib.dump(prediction_pipeline_no_bp, output_pipeline_path_no_bp)
    print(f"\nMAP 포함 예측 파이프라인을 '{output_pipeline_path_no_bp}' 파일로 성공적으로 저장했습니다.")
    print("이제 이 파일을 Flask 웹 서비스에서 MAP 기반 예측에 불러와 사용할 수 있습니다.")
except Exception as e:
    print(f"\n파이프라인 저장 중 오류 발생: {e}")
```
